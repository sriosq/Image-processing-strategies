```python
import matplotlib.pyplot as plt

# Data (manually extracted)
max_iters = [50, 70, 90, 130, 170, 210]
rmse = [2.5626390913189443,
        2.060247388590529,
        1.8689428280697271,
        1.7582552226691202,
        1.7162083747824954,
        1.7008376315271394]

# Plot
plt.figure(figsize=(6,5))
plt.plot(max_iters, rmse, marker='o')

# Labels and title
plt.xlabel("Maximum Iterations")
plt.ylabel("RMSE [Hz]")
plt.title("Effect of Max Iterations on RMSE\n(Tolerance = 0.1, PadSize = 14)")

# Grid for clarity
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
padsize = [40, 30, 20, 10, 14]
rmse = [1.773661133728459,
        1.7342616725289157,
        1.7093112022764019,
        1.7071191594417692,
        1.7055918552708615]

# Sort by padsize for nicer visualization
padsize_sorted, rmse_sorted = zip(*sorted(zip(padsize, rmse)))

# Plot
plt.figure(figsize=(6,5))
plt.plot(padsize_sorted, rmse_sorted, marker='o')

plt.xlabel("Pad Size")
plt.ylabel("RMSE")
plt.title("Effect of Pad Size on RMSE\n(Tolerance = 0.1, Max Iters = 200)")

plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
